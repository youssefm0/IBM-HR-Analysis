## Step 2: Data Cleaning

```python
#Drop unnecessary columns

df.drop(['EmployeeCount','Over18','StandardHours'] , axis = 1 , inplace = True)
```

```python
#Remove duplicates

df.drop_duplicates(inplace = True)
```

```python
#Change data type

df['Attrition'] = df['Attrition'].astype('category')
```

## Replace Unclear Columns 

```python
#Replace unclear values in 'Education' column with descriptive labels

mapping_education = {
           1:'High School',
           2:'Associates Degree',
           3:'Bachelors Degree',
           4:'Masters Degree',
           5:'Doctorate'
          }
df['Education'] = df['Education'].map(mapping_education)
```

```python
#Replace unclear values in 'EnvironmentSatisfaction' column with descriptive labels

mapping_EnvironmentSatisfaction = {
           1: 'Low',
           2: 'Medium',
           3: 'High',
           4: 'Very High'
            }
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(mapping_EnvironmentSatisfaction)
```

```python
#Replace unclear values in 'PerformanceRating' column with descriptive labels

mapping_PerformanceRating = {
           1: 'Low',
           2: 'Medium',
           3: 'High',
           4: 'Very High'
            }
df['PerformanceRating'] = df['PerformanceRating'].map(mapping_PerformanceRating)
```

```python
#Replace unclear values in 'JobSatisfaction' column with descriptive labels

mapping_JobSatisfaction = {
           1: 'Low',
           2: 'Medium',
           3: 'High',
           4: 'Very High'
            }
df['JobSatisfaction'] = df['JobSatisfaction'].map(mapping_JobSatisfaction)
```

```python
#Replace unclear values in 'RelationshipSatisfaction' column with descriptive labels 

mapping_RelationshipSatisfaction = {
           1: 'Low',
           2: 'Medium',
           3: 'High',
           4: 'Very High'
            }
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(mapping_RelationshipSatisfaction)
```

```python
#Replace unclear values in 'RelationshipSatisfaction' column with descriptive labels 

mapping_RelationshipSatisfaction = {
           0: 'None',
           1: 'Low',
           2: 'Medium',
           3: 'High'
}
df['StockOptionLevel'] = df['StockOptionLevel'].map(mapping_RelationshipSatisfaction)
```

```python
#Replace unclear values in 'RelationshipSatisfaction' column with descriptive labels 

mapping_RelationshipSatisfaction = {
          0: 'None',
          1: 'Low',
          2: 'Medium',
          3: 'High'
}
df['StockOptionLevel'] = df['StockOptionLevel'].map(mapping_RelationshipSatisfaction)
```

```python
#Replace unclear values in 'WorkLifeBalance' column with descriptive labels 

mapping_WorkLifeBalance = {
          1: 'Bad',
          2: 'Good',
          3: 'Better',
          4: 'Best'
}
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(mapping_WorkLifeBalance)
```

```python
#Add 'KM' to 'DistanceFromHome' for clarity

df['DistanceFromHome'] = df['DistanceFromHome'].astype(str) + ' KM'
```

## Step 3: Add A New Columns

```python
# check if there any employee younger than 20 or older than 80 to categorize the employees 

if (df['Age'] < 20).any() or (df['Age'] > 80).any():
    print(f'There Is Employees Bigger Than 80 Or Less Than 20')
else:
    print("There Is No Employees Bigger Than 80 Or Less Than 20")
```

```python
# check if there any employee younger than 20 or older than 80 to categorize the employees 

filtered_df = df[(df['Age'] < 20) | (df['Age'] > 80)]
print(filtered_df[['Age']])
```

## Step 4: Visualizations

```python
#Visualization 1: Employee Distribution by Department

df['Department'].value_counts().plot(kind = 'bar' , color = 'skyblue')
plt.title('Employee Distribution by Department')  # Title of the plot
plt.xlabel('Department')  # X-axis label
plt.ylabel('Number of Employees') # y-axis label
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid lines for better visualization
plt.show()
```

```python
#Visualization 2: Age Distribution of Employees

df['Age'].plot(kind = 'hist' , bins = 10 ,color = 'skyblue' , edgecolor = 'black')
plt.title('Age Distribution of Employees')  # Title of the plot
plt.xlabel('Age')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.figure(figsize=(10, 6))   # Set the figure size
plt.show()
```

```python
#Visualization 3: Average Monthly Income by Age Group


df.groupby('Age Group')['MonthlyIncome'].mean().reset_index().plot(x='Age Group', y='MonthlyIncome', marker='o', color='blue', linestyle='-')
plt.title('Average Monthly Income by Age Group')  # Title of the plot
plt.xlabel('Age Group')  # X-axis label
plt.ylabel('Average Monthly Income')  # Y-axis label
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid lines for better visualization
plt.tight_layout()  # Adjust layout to prevent overlap
plt.figure(figsize=(10, 6))
plt.show()
```

```python
#Visualization 4: Attrition Rate (Pie Chart)

plt.pie(
    df['Attrition'].value_counts(),  # Values for the pie chart
    labels=df['Attrition'].value_counts().index,  # Use 'Yes' and 'No' as labels
    autopct='%1.1f%%',  # Display percentage with one decimal place
    startangle=90,  # Start angle of the pie chart
    colors=['red', 'lightskyblue']  # Colors for the slices
)
plt.title('Attrition Rate')  # Title of the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
